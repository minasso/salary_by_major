```python
file1 = 'degrees-that-pay-back.csv'
file2 = 'salaries-by-college-type.csv'
file3 = 'salaries-by-region.csv'
```

```python
df = pd.read_csv(file1)
```

```python
df
```

Fix data types

```python
df.info()
```

```python
cols = df.columns
```

```python
for col in cols:
    print(col) 
    try:
        df[col] = df[col].apply(lambda x: x.replace('$','').replace(',',''))
        df[col] = pd.to_numeric(df[col])
    except:
        pass
```

```python
df.info()
```

Find highest starting salaries

```python
top_10_start = df.sort_values('Starting Median Salary',ascending=False).head(10)
```

```python
top_10_med = df.sort_values('Mid-Career Median Salary',ascending=False).head(10)
```

```python
top_10_pct = df.sort_values('Percent change from Starting to Mid-Career Salary',ascending=False).head(10)
```

Start doing some plotting

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,6
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
top_10_med
```

```python
sns.barplot(x='Undergraduate Major',y='Starting Median Salary',data = top_10_start)
plt.xticks(rotation=20)
plt.title('Starting Median Salary by Undergraduate Major')
```

```python
sns.barplot(x=top_10_med['Undergraduate Major'],y=top_10_med['Mid-Career Median Salary'])
plt.xticks(rotation=25)
plt.title('Mid-Career Median Salary by Undergraduate Major')
```

```python
sns.barplot(x='Undergraduate Major',y='Percent change from Starting to Mid-Career Salary',data = top_10_pct)
plt.xticks(rotation=20)
plt.title('Percent change from Starting to Mid-Career Salary by Undergraduate Major')
```
